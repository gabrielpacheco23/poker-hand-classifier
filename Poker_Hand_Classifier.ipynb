```python
import pandas as pd

data = pd.read_csv('/content/poker-hand-training-true.data')
data.head()
```

```python
X = data.drop('class', axis=1)
y = data['class']
```

```python
test_data = pd.read_csv('/content/poker-hand-testing.data')
# test_data = test_data.sample(50_000)
test_data.head()
```

```python
X_test, y_test = test_data.drop('class', axis=1), test_data['class']
```

```python
X_test.shape
```

### Using MLP

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(50, 40, 10),
                      solver='adam',
                      activation='relu',
                      alpha= 0.001)
```

```python
# model.fit(X_train, y_train)
model.fit(X, y)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, pred)
```
